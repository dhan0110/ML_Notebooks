```python
# Import the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

1. Histogram Plot

Distribution of data points. Univariate. Visual representation of bins

```python
# Generate a set of data points
Ndata = np.random.normal(100, 150, 1000)

# Plotting a basic histogram
#plt.hist(Ndata, bins=30)

plt.hist(Ndata, bins=30, color='green', edgecolor='black')

# Adding labels and title
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Basic Histogram')

# Display the plot
plt.show()
```

```python
Wdata = [89.2, 57.2, 63.4, 84.6, 90.2, 60.3, 88.7, 65.2, 79.6, 80.2, 93.5, 79.3, 72.5, 59.2, 77.2, 67.0, 88.2, 73.5]
plt.figure(figsize=(10,3))
sns.histplot(Wdata, bins=10)
plt.xlabel('Weights');
```

2. Box Plot

Univariate analysis. Predominantly used to identify outliers

```python
plt.boxplot(Wdata)
plt.xlabel('Weight')
plt.ylabel('Value in Kgs');
```

```python
sns.boxplot(Ndata)
plt.xlabel('Data')
plt.ylabel('Values');
```

3. Pie Chart

Univariate Analysis. Numerical proportions - Pie slices correspond to this.

```python
# Winter fruit sales
sales = [20, 30, 40, 10]
fruits = ['Apple', 'Orange', 'Strawberry', 'Others']
plt.pie(sales, labels = fruits);
```

```python
plt.pie(sales, labels=fruits, autopct = '%1.3f%%')
plt.title('Fruit Sales in Winter');
```

```python
df.content_rating.value_counts()
```

```python
plt.pie(df['content_rating'].value_counts(), labels = df.content_rating.unique());
```

```python
df.content_rating.value_counts().index
```

```python
plt.pie(df.content_rating.value_counts(), labels=df.content_rating.value_counts().index )
plt.show()
```

5. Donut / Doughnut Chart

A modification to the pie chart

```python
plt.pie(sales, labels=fruits, autopct = '%1.1f%%')
centre_circle = plt.Circle((0,0),0.80,fc='white')
fig = plt.gcf()
fig.gca().add_patch(centre_circle);
```

```python
explode=[0.05,0.05,0.05, 0.05]
plt.pie(sales, labels=fruits, explode=explode)
centre_circle = plt.Circle((0,0),0.80,fc='white')
fig = plt.gcf()
fig.gca().add_patch(centre_circle);
```

5. Line Plot

Continuous data, Bivariate analysis

```python
x = np.asarray([1, 2, 3, 4, 5, 6, 7, 8])
plt.plot(x,x);
```

```python
plt.plot(x, x+1, 'r+-')
plt.plot(x, x**2, 'g')
plt.plot(x, x**(1/2),'b*-');
```

6. Scatter plot (continuous vs continuous plot) \
Bivariate analysis

```python
plt.scatter(tips['total_bill'], tips['tip'], color='green', edgecolor='white')
plt.xlabel('Total Bill')
plt.ylabel('Tips')
plt.title('Total Bill v/s Tips');
```

```python
sns.scatterplot(data=tips,x='total_bill', y='tip', marker='X')
plt.plot(tips['total_bill'], tips['total_bill']*0.15, 'r--');
```

7. Bar Plot \
Distribution of categorical data. Univariate analysis

```python
plt.bar(tips.time.unique(), tips.time.value_counts(), color=['green', 'red'], label=tips.time.unique())
plt.xlabel('Meal Time')
plt.ylabel('No. of Orders')
plt.legend();
```

```python
sns.barplot(x=tips.time.unique(), y=tips.time.value_counts());
```

```python
sns.countplot(tips,x='time', hue='time');
```

8. Plot using Pandas

Pandas also supports plotting of data using the plot function on a DataFrame.

1. Stacked Bar Plot \
   Bars divided into a number of sub-bars to visualize the values of multiple variables at once

```python
df.plot(kind='bar',stacked='True');
```

```python
df.plot(kind='barh', stacked='True', color=['darkblue','lightgreen'])
plt.title('Total Orders per Day')
plt.xlabel('Order Counts');
```

2. Grouped Bar Plots

```python
df.plot(kind='bar',color=['lightblue','red']);
```

```python
tipsnew = tips[['time','size']].groupby('time', observed=False).value_counts().reset_index(name='count')
tipsnew
#sns.catplot(kind='bar', x='size',y='count', col='time', data=tipsnew);
```

```python
pivoted = tipsnew.pivot(index='size',columns='time',values='count')
#pivoted
pivoted.plot(kind='bar');
```
