1. Duplicate Data Handling

```python
df1 = pd.DataFrame(
                    {  'Name':['Jane', 'Jane', 'Aaron', 'Penelope', 'Jaane', 'Nicky',
                         'Armour', 'Ponting'] ,
                        'Age': [30, 30, 22, 40, 20, 30, 20, 25],
                     'Height': [165, 165, 120, 80, 162, 72, 124, 81],
                       'Score': [4.6, 4.6, 9.0, 3.3, 4, 8, 9, 3],
                       'State': ['NY', 'NY', 'FL', 'AL', 'NY', 'TX', 'FL', 'AL']
                        })

print("\n -------- Data with Duplicate Rows ----------- \n")
df1
```

```python
df1.duplicated()
```

2. Missing Data Handling

```python
df2 = pd.DataFrame(
                    {  'Name':[ 'Jane', 'Aaron', 'Penelope', 'Jaane', 'Nicky',
                         'Armour', 'Ponting'] ,
                        'Age': [ np.nan, 22, 40, 20, 30, 20, 25],
                     'Height': [ 165, 120, 80, 162, 72, 124, np.nan],
                       'Score': [ 4.6, 9.0, np.nan, 4, 8, 9, np.nan],
                       'State': [ 'NY', 'FL', None , 'NY', 'TX', 'FL', 'AL']
                        })
df2
```

```python
df2.isna()
```

```python
df2.isna().sum()
```

```python
df2.isna().sum(axis=1)
```

```python
df2[df2.isna().sum(axis=1)>=2].index           # Retrieving the index of the rows
```

```python
df2.drop([2,6])            ##similar to this df2.dropna(thresh=4)
```

```python
# Drop all the rows with missing values
df2.dropna()
```

```python
# Drop all the columns with missing values
df2.dropna(axis='columns')
```

```python
df2.dropna(thresh=4)
```

```python
df2
```

```python
np.mean(df2.Age)
```

```python
np.nanmedian(df2.Age)
```

```python
print(np.mean(df2.Score), np.nanmedian(df2.Score))
```

```python
AgeFiller = int(np.mean(df2.Age))
ScoreFiller = np.nanmedian(df2.Score)

df2.fillna(value={'Age':AgeFiller, 'Score':ScoreFiller})
```
